```python
from keras.models import model_from_json
import numpy as np
from PIL import Image
import keyboard
import time
from mss import mss
```

```python
mon = {"top":455, "left":735, "width":265, "height":130} #ekranin kullanilacak kismi
sct = mss()

width = 125
height = 50

# model yükle
model = model_from_json(open("model.json","r").read()) # modeli yükle
model.load_weights("trex_weights.weights.h5") # modelin agirliklarini yükle

# down = 0, right = 1, up = 2
labels = ["Down", "Right", "Up"] # modelin tahmin edebilecegi degerler

framerate_time = time.time() # fps hesaplamak icin
counter = 0
i = 0
delay = 0.4
key_down_pressed = False
while True:
    
    img = sct.grab(mon) # ekran goruntusunu al
    im = Image.frombytes("RGB", img.size, img.rgb) # goruntuyu RGB ye cevir
    im2 = np.array(im.convert("L").resize((width, height))) # griye cevir ve boyutlandir
    im2 = im2 / 255 # normalize et
    
    X =np.array([im2]) # modelin bekledigi sekilde boyutlandir
    X = X.reshape(X.shape[0], width, height, 1) # modelin bekledigi sekilde boyutlandir
    r = model.predict(X) # tahmin yap
    
    result = np.argmax(r) # tahminin en yuksek oldugu degeri al
    
    
    if result == 0: # down = 0
        
        keyboard.press(keyboard.KEY_DOWN)
        key_down_pressed = True
        
    elif result == 2:    # up = 2
        
        if key_down_pressed:
            keyboard.release(keyboard.KEY_DOWN)
        time.sleep(delay)
        keyboard.press(keyboard.KEY_UP)
        
        if i < 1500: 
            time.sleep(0.3) 
        elif 1500 < i and i < 5000:
            time.sleep(0.2)
        else:
            time.sleep(0.17)
            
        keyboard.press(keyboard.KEY_DOWN) # bir sonraki frame de tekrar bas
        keyboard.release(keyboard.KEY_DOWN) 
    
    counter += 1
    
    if (time.time() - framerate_time) > 1: #oynama suresini hesapla
        
        counter = 0
        framerate_time = time.time() 
        if i <= 1500:
            delay -= 0.003
        else:
            delay -= 0.005
        if delay < 0:
            delay = 0
            
        print("---------------------") 
        print("Down: {} \nRight:{} \nUp: {} \n".format(r[0][0],r[0][1],r[0][2]))
        i += 1
        
```
